```python
from math import pi, sqrt
import matplotlib.pyplot as plt
import numpy as np
```

```python
global R

# def plotf(res):
#     if res > 2

def f(x):
    res = (pi / 3) * (x ** 2) * (2 * R - x)
    return res
```

```python
def goldenSectionSearch (f, l, r, eps):
    phi = (1 + sqrt(5)) / 2
    resphi = 2 - phi
    x1 = l + resphi * (r - l)
    x2 = r - resphi * (r - l)
    f1 = f(x1)
    f2 = f(x2)
    while True:
        print(l, r, r - l, x1, x2, f(x1), f(x2))
        if f1 > f2:
            r = x2
            x2 = x1
            f2 = f1
            x1 = l + resphi * (r - l)
            f1 = f(x1)
        else:
            l = x1
            x1 = x2
            f1 = f2
            x2 = r - resphi * (r - l)
            f2 = f(x2)
        if abs(r - l) < eps:
            break

        print(l, r, r - l, x1, x2, f(x1), f(x2))
    return (x1 + x2) / 2
```

```python
R = 40
x = np.linspace(R, 2 * R, 100)
plt.plot(x, f(x))
plt.show()
```

```python
res = goldenSectionSearch(f, R, 2 * R, 0.1)
```
